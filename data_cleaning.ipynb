### Treating with missing values and float values

```python
print("NaN Ejercicio Fiscal",df['EJERCICIO FISCAL'].isnull().sum())
print("NaN PESO DESEMBARCADO KILOGRAMOS",df['PESO DESEMBARCADO KILOGRAMOS'].isnull().sum())
print("NaN PESO VIVO KILOGRAMOS",df['PESO VIVO KILOGRAMOS'].isnull().sum())
print("NaN VALOR PESOS",df['VALOR PESOS'].isnull().sum())
```

```python
#drop rows in which all data is missing and reset the index
df=df.dropna(how='all').reset_index(drop=True)
display(df)
```

```python
# transforming string or float values to numbers
cols = ['PESO VIVO KILOGRAMOS', 'PESO DESEMBARCADO KILOGRAMOS', 'VALOR PESOS','EJERCICIO FISCAL']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
#replace NaN with O values
df['EJERCICIO FISCAL'] = df['EJERCICIO FISCAL'].replace(np.nan, 0)
df['PESO DESEMBARCADO KILOGRAMOS'] = df['PESO DESEMBARCADO KILOGRAMOS'].replace(np.nan, 0)
df['PESO VIVO KILOGRAMOS']= df['PESO VIVO KILOGRAMOS'].replace(np.nan, 0)
df['VALOR PESOS'] = df['VALOR PESOS'].replace(np.nan, 0)
```

```python
print("NaN ejercicio fiscal",df['EJERCICIO FISCAL'].isnull().sum())
print("NaN PESO DESEMBARCADO KILOGRAMOS",df['PESO DESEMBARCADO KILOGRAMOS'].isnull().sum())
print("NaN PESO VIVO KILOGRAMOS",df['PESO VIVO KILOGRAMOS'].isnull().sum())
print("NaN VALOR PESOS",df['VALOR PESOS'].isnull().sum())
```

transform strings to float values and then to to intergers

```python
print("NaN ejercicio fiscal:",df['EJERCICIO FISCAL'].isnull().values.any())
print("NaN PESO DESEMBARCADO KILOGRAMOS:",df['PESO DESEMBARCADO KILOGRAMOS'].isnull().values.any())
print("NaN PESO VIVO KILOGRAMOS:",df['PESO VIVO KILOGRAMOS'].isnull().values.any())
print("NaN VALOR PESOS:",df['VALOR PESOS'].isnull().values.any())
```

```python
#transform column 'FECHA' from float to intergers
columns=['PESO VIVO KILOGRAMOS', 'PESO DESEMBARCADO KILOGRAMOS', 'VALOR PESOS']
df[columns]=df[columns].astype(np.int64)
```

```python
columnx=['EJERCICIO FISCAL','ENTIDAD FEDERATIVA','NOMBRE OFICINA', 'MES DE CORTE', 'ORIGEN','NOMBRE COMUN','NOMBRE PRINCIPAL']
df[columnx] = df[columnx].astype(str)
```

```python
df.info()
```

```python
#rename columns and data frame and deletion of the 2 last columns which will not be used. 
df = df.rename(columns={'ENTIDAD FEDERATIVA':'ESTADO','EJERCICIO FISCAL':'FECHA','NOMBRE OFICINA':'OFICINA',"NOMBRE PRINCIPAL":'FAMILIA',"NOMBRE COMUN":'ESPECIE'}).drop(columns=['CLAVE DE ENTIDAD','CLAVE DE OFICINA'])
df.head(10)
```
